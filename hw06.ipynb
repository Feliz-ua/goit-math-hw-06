**Завдання 1. Оптимальна ціна товару** 

Маємо:\
Попит на навушники залежить від ціни за формулою:\
*Q(x)=2000-0.8x*, lt *x*-ціна в гривнях, *Q(x)*-кількість проданих одиниць на місяць.

Собівартість одних навушників становить 800 грн. Тоді прибуток компанії:
*p(x}=(x-800)\cdot Q(x)=(x-800)(2000-0.8x)*
